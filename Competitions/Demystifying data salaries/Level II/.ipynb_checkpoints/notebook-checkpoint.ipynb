## ðŸ’¾ **Data**

The data comes from a survey hosted by an HR consultancy, available in `'salaries.csv'`.

#### Each row represents a single employee's salary record for a given year:
- **`work_year`** - The year the salary was paid.  
- **`experience_level`** - Employee experience level:  
  - **`EN`**: Entry-level / Junior  
  - **`MI`**: Mid-level / Intermediate  
  - **`SE`**: Senior / Expert  
  - **`EX`**: Executive / Director  
- **`employment_type`** - Employment type:  
  - **`PT`**: Part-time  
  - **`FT`**: Full-time  
  - **`CT`**: Contract  
  - **`FL`**: Freelance  
- **`job_title`** - The job title during the year.  
- **`salary`** - Gross salary paid (in local currency).  
- **`salary_currency`** - Salary currency (ISO 4217 code).  
- **`salary_in_usd`** - Salary converted to USD using average yearly FX rate.  
- **`employee_residence`** - Employee's primary country of residence (ISO 3166 code).  
- **`remote_ratio`** - Percentage of remote work:  
  - **`0`**: No remote work (<20%)  
  - **`50`**: Hybrid (50%)  
  - **`100`**: Fully remote (>80%)  
- **`company_location`** - Employer's main office location (ISO 3166 code).  
- **`company_size`** - Company size:  
  - **`S`**: Small (<50 employees)  
  - **`M`**: Medium (50â€“250 employees)  
  - **`L`**: Large (>250 employees)  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/salaries.csv')
df.head()
```

## Create a bar chart displaying the top 5 job titles with the highest average salary (in USD).

```python
# Horizontal bar chart for top 5 job titles with the highest average salary (in USD)
plt.figure(figsize=(10, 6))
top_job_titles = df.groupby('job_title')['salary_in_usd'].mean().nlargest(5)
top_job_titles[::-1].plot(kind='barh', color='skyblue')
plt.title('Top 5 Job Titles with Highest Average Salary (in USD)')
plt.xlabel('Average Salary (in USD)')
plt.ylabel('Job Title')
plt.tight_layout()
plt.show()
```

## Compare the average salaries for employees working remotely 100%, 50%, and 0%. What patterns or trends do you observe?

```python
remote_salary_comparison = df.groupby('remote_ratio')['salary_in_usd'].mean()
plt.figure(figsize=(8, 5))
remote_salary_comparison.plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Average Salary for Different Remote Ratios')
plt.xlabel('Remote Ratio (%)')
plt.ylabel('Average Salary (in USD)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Visualise the salary distribution (in USD) across company sizes (S, M, L). Which company size offers the highest average salary?

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='company_size', y='salary_in_usd', palette='Set3', hue = 'company_size', order=['S','M','L'])
plt.title('Salary Distribution Across Company Sizes (in USD)')
plt.xlabel('Company Size')
plt.ylabel('Salary (in USD)')
plt.tight_layout()
plt.show()
```

```python
# Task 3: Calculate the highest average salary by company size
avg_salary_by_size = df.groupby('company_size')['salary_in_usd'].mean()
companies = avg_salary_by_size.index
salaries = avg_salary_by_size.values

for i in range(len(companies)):
    print(f"{companies[i]}: ${salaries[i]:,.0f}")
```

```python
highest_avg_salary_size = avg_salary_by_size.idxmax()
highest_avg_salary = avg_salary_by_size.max()
print(f"{companies[np.argmax(salaries)]} companies offer the highest average salaries")
```
