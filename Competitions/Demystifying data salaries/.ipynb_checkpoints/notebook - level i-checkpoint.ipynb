## ðŸ’¾ **Data**

The data comes from a survey hosted by an HR consultancy, available in `'salaries.csv'`.

#### Each row represents a single employee's salary record for a given year:
- **`work_year`** - The year the salary was paid.  
- **`experience_level`** - Employee experience level:  
  - **`EN`**: Entry-level / Junior  
  - **`MI`**: Mid-level / Intermediate  
  - **`SE`**: Senior / Expert  
  - **`EX`**: Executive / Director  
- **`employment_type`** - Employment type:  
  - **`PT`**: Part-time  
  - **`FT`**: Full-time  
  - **`CT`**: Contract  
  - **`FL`**: Freelance  
- **`job_title`** - The job title during the year.  
- **`salary`** - Gross salary paid (in local currency).  
- **`salary_currency`** - Salary currency (ISO 4217 code).  
- **`salary_in_usd`** - Salary converted to USD using average yearly FX rate.  
- **`employee_residence`** - Employee's primary country of residence (ISO 3166 code).  
- **`remote_ratio`** - Percentage of remote work:  
  - **`0`**: No remote work (<20%)  
  - **`50`**: Hybrid (50%)  
  - **`100`**: Fully remote (>80%)  
- **`company_location`** - Employer's main office location (ISO 3166 code).  
- **`company_size`** - Company size:  
  - **`S`**: Small (<50 employees)  
  - **`M`**: Medium (50â€“250 employees)  
  - **`L`**: Large (>250 employees)

## Check numerical data distributions

```python
ignore_col = ['salary']
num_cols = list(df.select_dtypes(exclude='object').columns)
num_cols = [c for c in num_cols if c not in ignore_col]

for col in num_cols:
    plt.figure(figsize = (15, 6))
    plt.subplot(1, 2, 1)
    sns.boxplot(data=df, y=col)
    plt.title(col)
    plt.subplot(1, 2, 2)
    sns.histplot(data=df, x=col, kde=True)
    plt.title(col)
    plt.show()
```

```python
cat_cols = list(df.select_dtypes(include='object').columns)

for col in cat_cols:
    plt.figure(figsize=(15, 6))
    sns.countplot(x=col, data=df)
    plt.title(f'Category Count for {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)  # Rotate labels if necessary
    plt.show()    
```

## How many records are in the dataset, and what is the range of years covered?

```python
num_records = df.shape[0]
year_range = df['work_year'].min(), df['work_year'].max()
print(f"Record count: {num_records:,.0f} records")
print(f"Year range: {year_range}")
```

## What is the average salary (in USD) for Data Scientists and Data Engineers? Which role earns more on average?

```python
roles_of_interest = ['Data Scientist', 'Data Engineer']
avg_salary_roles = df[df['job_title'].isin(roles_of_interest)].groupby('job_title')['salary_in_usd'].mean().round(0)
roles = avg_salary_roles.index
salaries = avg_salary_roles.values
print("Salary list")
print("---------------")
for i in range(len(avg_salary_roles.values)):
    print(f"{roles[i]}: ${salaries[i]:,.0f}")
print()
print(f"On average, {roles[np.argmax(salaries)]}s earn more...")
```

## How many full-time employees based in the US work 100% remotely?

```python
full_time_remote_us = df[(df['employment_type'] == 'FT') & 
                         (df['employee_residence'] == 'US') & 
                         (df['remote_ratio'] == 100)].shape[0]
print(f"{full_time_remote_us:,.0f} employees")
```

## What is the average salary by job title?

```python
top = 20

plt.figure(figsize = (16, 6))

df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending = False)[:top][::-1].plot(kind = 'barh')
plt.title(f"Avg Salaries by Job Title, Top {top}")
plt.xlabel("Avg. salary")
plt.tight_layout()
```
